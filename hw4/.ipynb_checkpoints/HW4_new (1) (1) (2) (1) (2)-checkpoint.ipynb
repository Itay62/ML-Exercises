````python
# Function for ploting the decision boundaries of a model
# You will use it later
def plot_decision_regions(X, y, classifier, resolution=0.01):

    # setup marker generator and color map
    markers = ('.', '.')
    colors = ('blue', 'red')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')
````

## Logistic Regression

(15 Points)

Implement the Logistic Regression algorithm that uses gradient descent for finding the optimal theta vector. 

Where:
$$
h_\theta(x) = \frac{1}{1 + e^{-\theta^T x}}
$$

$$
J(\theta)=\frac{1}{m} \sum_{d=1}^{m} - y^{(d)}ln(h_\theta(x^{(d)}) - (1 - y^{(d)})ln(1 - h_\theta(x^{(d)})
$$

Your class should contain the following functions:
1. fit - the learning function
1. predict - the function for predicting an instance after the fit function was executed

\* You can add more functions if you think this is necessary

Your model should also store a list of the costs that you've calculated in each iteration

````python
def bias_trick(X):
    ones = np.ones(X.shape[0], dtype=int)
    return np.column_stack((ones,X))
````

````python
class LogisticRegressionGD(object):
    """
    Logistic Regression Classifier using gradient descent.

    Parameters
    ------------
    eta : float
      Learning rate (between 0.0 and 1.0)
    n_iter : int
      Passes over the training dataset.
    eps : float
      minimal change in the cost to declare convergence
    random_state : int
      Random number generator seed for random weight
      initialization.
    """
    
    def __init__(self, eta=0.00005, n_iter=10000, eps=0.000001, random_state=42):
        self.eta = eta
        self.n_iter = n_iter
        self.eps = eps
        self.random_state = random_state
        self.theta = []
        self.J_history = []

    def fit(self, X, y):
        """ 
        Fit training data (the learning phase).
        Updating the theta vector in each iteration using gradient descent.
        Store the theta vector in an attribute of the LogisticRegressionGD object.
        Stop the function when the difference between the previous cost and the current is less than eps
        or when you reach n_iter.

        Parameters
        ----------
        X : {array-like}, shape = [n_examples, n_features]
          Training vectors, where n_examples is the number of examples and
          n_features is the number of features.
        y : array-like, shape = [n_examples]
          Target values.

        """
        X = bias_trick(X)
        np.random.seed(self.random_state)
        self.theta = np.random.random(X.shape[1])
        m = X.shape[0]
        for i in range(self.n_iter):
            cost = self.compute_cost(X, y, self.theta)
            if  i > 0 and (abs(self.J_history[-1] - cost) < self.eps):
                break    
            self.J_history.append(cost)
            h = self.hipo(X, self.theta)
            self.theta -= ((self.eta)*(X.T.dot(h-y)))  
        return self.theta, self.J_history
        
    def predict(self, X):
        """Return the predicted class label"""
        m = X.shape[0]
        predicts = []
        X = bias_trick(X)
   
        predicts = [ 1 if self.hipo(X[i], self.theta) > 0.5 else 0 for i in range(m) ]
        
        
        return np.array(predicts)

    
    def hipo(self, X, theta):
        theta_t = np.transpose(theta)
        dot_X_theta = np.dot(X, theta_t)
        e = np.exp(dot_X_theta)
        hipo = np.divide(e, 1 + e)
        return hipo
            
    def compute_cost(self, X, y, theta):
        J = 0
        h = self.hipo(X, theta)
        m = np.shape(np.array(X))[0]
        J = (np.sum( (-y * np.log(h) ) - 
                            ( (1 - y) * np.log(1 - h) ) ) ) / m
        return J
````

````python
def compute_accuracy_logistic(validation_set, logisticRegressionGD):
    """
    Compute the accuracy of a given a testset using a LogisticRegressionGD object.
    
    Input
        - testset: The testset for which to compute the accuracy (Numpy array). compoed by class labels
        - logisticRegressionGD : A logisticRegressionGD object capable of prediciting the class for each instance in the testset.
        
    Ouput
        - Accuracy = #Correctly Classified / #testset size
    """
    correct_count = 0
    validation_set_x, validation_set_y = validation_set[:,:2], validation_set[:,-1]
    #validation_set_x = bias_trick(validation_set_x)
    for i in range(len(validation_set_x)):
        if logisticRegressionGD.predict(validation_set_x)[i] == validation_set_y[i]: # if prediction is correct increase counter
            correct_count = correct_count + 1
    return  correct_count / len(validation_set)
````

## Normal distribution pdf

(2 Points)

Implement the normal distribution pdf 
$$
f(x;\mu,\sigma) = \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{1}{2}\big{(}\frac{x-\mu}{\sigma}\big{)}^2}
$$
Write an efficient vectorized code

````python
# calc normal pdf    
def norm_pdf(data, mu, sigma):
  
    x_minus_std_square = (np.subtract(data, mu)) ** 2
    std_square_times_2 = np.multiply(sigma ** 2,2)
    power = -np.divide(x_minus_std_square, std_square_times_2)
    right = np.exp(power)
    denominator = np.multiply(std_square_times_2, np.pi)
    left = np.divide(1, np.sqrt(denominator))
    result = np.multiply(left, right)
    
    return result
````

## Expectation Maximization

(15 Points)

Implement the Expectation Maximization algorithm for gaussian mixture model.

The class should hold the distribution params.

Use -log likelihood as the cost function:
$$
cost(x) = \sum_{d=1}^{m}-log(w * pdf(x; \mu, \sigma))
$$

\* The above is the cost of one gaussian. Think how to use the cost function for gaussian mixture.

Your class should contain the following functions:
1. init_params - initialize distribution params
1. expectation - calculating responsibilities
1. maximization - updating distribution params
1. fit - the learning function
1. get_dist_params - return the distribution params

\* You can add more functions if you think this is necessary

Don't change the eps params (eps=0.01)

When you need to calculate the pdf of a normal distribution use the function `norm_pdf` that you implemented above.

````python
class EM(object):
    """
    Naive Bayes Classifier using Gauusian Mixture Model (EM) for calculating the likelihood.

    Parameters
    ------------
    k : int
      Number of gaussians in each dimension
    n_iter : int
      Passes over the training dataset in the EM proccess
    eps: float
      minimal change in the cost to declare convergence
    random_state : int
      Random number generator seed for random params initialization.
    """
    
    def __init__(self, k=2, n_iter=1000, eps=0.01,random_state=42):
        self.k = k
        self.n_iter = n_iter
        self.eps = eps
        self.random_state=random_state
        self.mu = []
        self.sigma = []
        self.weight = []
        self.responsibilities = [[]] * self.k

        
    # initial guesses for parameters
    def init_params(self, data):
        """
        Initialize distribution params
        """
        np.random.seed(self.random_state)
        self.mu = np.random.random(self.k)
        self.sigma = np.random.random(self.k)
        self.weight = np.ones((self.k, 1), dtype=np.float32) / self.k
        #self.responsibilities = np.ones([len(data), self.k])
     
        
    def expectation(self, data):
        """
        E step - calculating responsibilities
        """
        sum_responsibilities = 0
        self.responsibilities = [[]] * self.k
        for k in range(self.k):
            sum_responsibilities += self.weight[k] * norm_pdf(data, self.mu[k], self.sigma[k])
        for k in range(self.k):
            self.responsibilities[k] = self.weight[k] * norm_pdf(data, self.mu[k], self.sigma[k]) / sum_responsibilities
            
        return self.responsibilities
    
    
    def maximization(self, data):
        """
        M step - updating distribution params
        """
        length_of_data = len(data)
        for j in range(self.k):
        
            self.weight[j] = np.divide(np.nansum(self.responsibilities[j]), length_of_data) ## new weight
            self.mu[j] = np.divide(np.dot((self.responsibilities[j]), data), length_of_data * self.weight[j])  ## new mean 
            self.sigma[j] = np.sqrt( np.divide(np.dot( (self.responsibilities[j]), np.square(np.subtract(data, self.mu[j]))), length_of_data * self.weight[j])) ## new sigma
       
    def fit(self, data):
        """ 
        Fit training data (the learning phase).
        Use init_params and then expectation and maximization function in order to find params 
        for the distribution. 
        Store the params in attributes of the EM object.
        Stop the function when the difference between the previous cost and the current is less than eps
        or when you reach n_iter.
        """
        self.init_params(data)
        cost_history = []
        cost_history.append(self.cost(data))
        
        for i in range(1, self.n_iter):
            #training the data
            self.expectation(data)
            self.maximization(data)
            
            #checking costs
            cost_history.append(self.cost(data))
            if (abs(cost_history[i] - cost_history[i-1]) ) < self.eps:
               
                break;
    
    def get_dist_params(self):
        return self.weight, self.mu, self.sigma 
    
    def cost(self, data):
        cost = 0
        for k in range(self.k):
            helper = 0
            for i in range(len(data)):
                helper += self.weight[k] * norm_pdf(data[i], self.mu[k], self.sigma[k])
            cost += -np.log(helper)
        return cost
````

## Naive Bayes

(15 Points)

Implement the Naive Bayes algorithm.

For calculating the likelihood use the EM algorithm that you implemented above to find the distribution params. With these params you can calculate the likelihood probability.

Calculate the prior probability directly from the training set.

Your class should contain the following functions:
1. fit - the learning function
1. predict - the function for predicting an instance (or instances) after the fit function was executed

\* You can add more functions if you think this is necessary


````python
class NaiveBayesGaussian(object):
    """
    Naive Bayes Classifier using Gauusian Mixture Model (EM) for calculating the likelihood.

    Parameters
    ------------
    k : int
      Number of gaussians in each dimension
    random_state : int
      Random number generator seed for random params initialization.
    """
    
     
    def __init__(self, k=2):
        self.k = k
        self.params_dict = {} #(label,feature index):(weight, mu, sigma)
        self.posteriors = []
        self.labels = []
        self.count = []
        self.len_data = 0
        
    def fit(self, X,y):
        """ 
        Fit training data.

        Parameters
        ----------
        X : array-like, shape = [n_examples, n_features]
          Training vectors, where n_examples is the number of examples and
          n_features is the number of features.
        y : array-like, shape = [n_examples]
          Target values.
        """
        
        self.len_data = len(X)
        self.labels,self.count = np.unique(y,return_counts=True)
        for label in self.labels:
            for i in range(np.shape(X)[1]):
                feature_data = X[:,i] #column feature
                feature_and_class = feature_data[np.where(y == label)]
                em = EM()
                em.fit(feature_and_class)
                weight, mu, sigma = em.get_dist_params()
                self.params_dict[(label,i)] = (weight, mu, sigma)
                
        return self.params_dict
        
    
    def get_instance_likelihood_for_label(self,X,label):
       
        for i in range(len(X)):
            pdf = 0
            likelihood = 1
            params = self.params_dict.get((label,i))
            for j in range(len(params[0])):
                w = params[0][j]
                mu = params[1][j]
                sigma = params[2][j]
                normal_pdf = norm_pdf(X[i], mu, sigma)
                pdf += w*normal_pdf
            likelihood *=pdf     
                    
        return likelihood
            
    def get_posteriors(self,X):
        
        i = 0
        for label in self.labels:
            likelihood = 0 #likelihoods for each label
            prior = 0
            likelihood = self.get_instance_likelihood_for_label(X,label)
            prior = self.count[i]/self.len_data
            i += 1
            self.posteriors.append(likelihood*prior)
            
        return self.posteriors
    
    def predict(self, X):
        """Return the predicted class label"""
        predict = []
        for instance in range(len(X)):
            self.posteriors = []
            self.posteriors = self.get_posteriors(X[instance])
            max_posteriors = max(self.posteriors) 
            index_max_posteriors = self.posteriors.index(max_posteriors)
            self.labels[index_max_posteriors]
            predict.append(self.labels[index_max_posteriors])
        return predict
````

````python
def compute_accuracy_naive(validation_set, NaiveBayesGaussian):
    """
    Compute the accuracy of a given a testset using a NaiveBayesGaussian object.
    
    Input
        - testset: The testset for which to compute the accuracy (Numpy array). compoed by class labels
        - NaiveBayesGaussian : A NaiveBayesGaussian object capable of prediciting the class for each instance in the testset.
        
    Ouput
        - Accuracy = #Correctly Classified / #testset size
    """
    correct_count = 0
    validation_set_x, validation_set_y = validation_set[:,:2], validation_set[:,-1]
    for i in range(len(validation_set_x)):
        if NaiveBayesGaussian.predict(validation_set_x)[i] == validation_set_y[i]: # if prediction is correct increase counter
            correct_count = correct_count + 1
    return  correct_count / len(validation_set)
````
